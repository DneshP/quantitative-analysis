## Financial Instruments

### Stocks

- **Definition:**
  - Equity representing ownership in a company.

- **Key Points:**
  - Share ownership, profit-sharing through price increase or dividends.

### Bonds

- **Definition:**
  - Fixed-income debt with fixed-rate coupon or variable interest payments.

- **Key Points:**
  - Fixed duration loan, regular interest payments, principal repayment.

### Annuities

- **Definition:**
  - Insurance contracts providing future fixed-income streams.

- **Key Points:**
  - Purchased for retirement, guaranteed payments for a period or life.

### Cash and Equivalents

- **Definition:**
  - Highly liquid, low-risk, short-term investments.

- **Key Points:**
  - Bank accounts, treasury bills, money market funds, easily accessible.

### Commodities

- **Definition:**
  - Basic goods used in commerce, traded in spot or derivatives markets.

- **Key Points:**
  - Examples include gold, oil, natural gas.

### Futures

- **Definition:**
  - Legal agreements obliging the buyer to buy/sell an asset at a preset price and time.

- **Key Points:**
  - Used for hedging, settled daily, marked to market.

### Forward

- **Definition:**
  - Private, customizable OTC agreements similar to futures.

- **Key Points:**
  - Settled at agreement end, direct trading without a central exchange.

### Options

- **Definition:**
  - Financial derivatives giving the right (not obligation) to buy/sell an asset at a specific price by a certain date.

- **Key Points:**
  - Call (buy) and put (sell) options, used for hedging or speculation.

### Currencies

- **Definition:**
  - International currency trading in the forex market.

- **Key Points:**
  - Exchange one currency for another at market rates, speculation on currency values.

### ETFs (Exchange-Traded Funds)

- **Definition:**
  - Pooled investment securities traded like stocks.

- **Key Points:**
  - Basket of securities (stocks, bonds, commodities), intraday trading.

### REITs (Real Estate Investment Trusts)

- **Definition:**
  - Companies owning, operating, or financing income-generating real estate.

- **Key Points:**
  - Publicly traded, steady income from real estate without direct ownership.

### Mutual Funds

- **Definition:**
  - Pooled investment portfolios managed by professionals.

- **Key Points:**
  - Diversified portfolios, purchased at end of trading day, net asset value.

### Hedge Funds

- **Definition:**
  - Actively managed investment pools with diverse trading strategies.

- **Key Points:**
  - Aim for above-average returns, higher fees, often risky strategies.

**Long a Call:**

Scenario: You're optimistic about a stock's future.
Action: You buy a call option.
Right: To buy the underlying stock at a specified price (strike price).
Analogy: Like making a reservation to buy the stock at a set price.

**Short a Call:**

Scenario: You own the stock, not expecting a significant price rise.
Action: You sell a call option.
Obligation: Potentially sell your stock at a specified price if the buyer exercises.
Analogy: Offering others the opportunity to buy the stock from you at a certain price.

**Long a Put:**

Scenario: Concerned about a potential drop in a stock's price.
Action: You buy a put option.
Right: To sell the underlying stock at a specified price (strike price).
Analogy: Having the option to sell the stock at a set price if things go south.

**Short a Put:**

Scenario: Believing a stock will remain stable or go up.
Action: You sell a put option.
Obligation: Potentially buy the stock at a specified price if the buyer exercises.
Analogy: Giving someone else the opportunity to sell you the stock at a certain price.

**Anatomy of a Trade:**

1. **Acquisition of Information and Quotes:**
   - Making informed decisions by gathering relevant information and quotes.

2. **Routing of Order:**
   - Choosing the market and broker as the means to execute the order.

3. **Execution of Order:**
   - Matching the order between the buyer and seller according to the rules of the market.

4. **Confirmation, Clearance, and Settlement:**
   - Confirmation occurs at the end of order execution, ensuring accuracy.
   - Clearance involves recording and matching the trading records.
   - Settlement includes delivering the security and its payment.

**Market Makers:**
Market makers create liquidity by providing bid and ask quotes and matching orders. They make profits on the spread, the difference between the buying (bid) and selling (ask) prices.

**Scalping:**
Scalping is a short-term trading strategy often employed by day traders. Traders aim to make quick profits by buying at a lower price and selling at a higher price within a short time frame.

**Portfolio Rebalancing:**
The goal of portfolio rebalancing is to ensure that the portfolio's asset allocation aligns with the investor's objectives. It helps mitigate risk and maintain the desired balance between risk and reward over time.

# Electronic Market

The electronic market operates on a discrete price grid, with prices linearly arranged based on their magnitude. The price ladder progresses from low to high prices. In any given market, there is always a minimum tick size, representing the adjacent price difference of a financial instrument.

- Stocks generally trade in one-cent tick increments.
- Currencies deal in pips (percentage in point, price interest point).
- Rates deal in bps (basis points).

CDP (Central Depository Participant) acts as the counterparty to both the buyer and seller, assuming the associated risks of both parties.

## Agency Trading

In agency trading, the intermediary facilitates orders between clients without taking positions in the securities being traded. The agency's sole role is to execute client orders, earning commissions or fees for its services.

## Proprietary Trading

In proprietary trading, an entity uses its own capital for market transactions. Unlike agency trading, where the focus is on executing client orders, proprietary trading involves taking speculative positions to capitalize on market movements. Both types of trading are subject to different regulatory conditions.

## Held Order

A held order is where the broker has an obligation to fulfill a client's order.

## Market-Not-Held Order

In a market-not-held order, the institutional trader hires a broker-dealer to execute an order. This provides the broker-dealer with more discretion in the execution process.
