# Quantitative Trading Analysis

## Input Data Categories

- **Market States:**
  - Price movements (tick data)
  - Bid-ask spread in the limit order book (LOB)
  - Lot size

- **Financial News:**
  - Macroeconomics news
  - Analytical reports
  - Earnings call transaction transcripts

- **Fundamentals:**
  - Company's revenue, cash flow, Earnings per Share (EPS)

- **Technicals:**
  - Derived technical indicators from raw price series
    - Moving average
    - Stochastic indicators

## Terms

- **Tick:**
  - Increase or decrease in the respective security or bond

- **Bid-ask spread**

- **Limit Order Book (LOB)**

- **Lot Size:**
  - Minimum amount of security that can be traded

- **Earnings per Share (EPS)**

## Order Execution Process

- Based on financial indicators
- Purpose: Beat (alpha) and make profits

- Institutional algorithms aim at maintaining market stability.

- Liquidity is proportional to the asset's ease of movement.

- **Slippage:**
  - Example: (X + Î”) change in price after an order has been placed.

- **Dark Pools:**
  - Private exchanges where certain orders take place away from central stock exchanges.

- **Iceberg Orders:**
  - Breaking down large institutional orders into smaller orders to preserve market stability and reduce impact.

## Common Indicator for Short-Term Traders

- When the price moves from below to above volume-weighted average price, indicative to enter a long position.

## Arbitrage

- Practice of exploiting price differences of the same asset in different markets.

## Financial Instruments

### Stocks

- **Definition:**
  - Equity representing ownership in a company.

- **Key Points:**
  - Share ownership, profit-sharing through price increase or dividends.

### Bonds

- **Definition:**
  - Fixed-income debt with fixed-rate coupon or variable interest payments.

- **Key Points:**
  - Fixed duration loan, regular interest payments, principal repayment.

### Annuities

- **Definition:**
  - Insurance contracts providing future fixed-income streams.

- **Key Points:**
  - Purchased for retirement, guaranteed payments for a period or life.

### Cash and Equivalents

- **Definition:**
  - Highly liquid, low-risk, short-term investments.

- **Key Points:**
  - Bank accounts, treasury bills, money market funds, easily accessible.

### Commodities

- **Definition:**
  - Basic goods used in commerce, traded in spot or derivatives markets.

- **Key Points:**
  - Examples include gold, oil, natural gas.

### Futures

- **Definition:**
  - Legal agreements obliging the buyer to buy/sell an asset at a preset price and time.

- **Key Points:**
  - Used for hedging, settled daily, marked to market.

### Forward

- **Definition:**
  - Private, customizable OTC agreements similar to futures.

- **Key Points:**
  - Settled at agreement end, direct trading without a central exchange.

### Options

- **Definition:**
  - Financial derivatives giving the right (not obligation) to buy/sell an asset at a specific price by a certain date.

- **Key Points:**
  - Call (buy) and put (sell) options, used for hedging or speculation.

### Currencies

- **Definition:**
  - International currency trading in the forex market.

- **Key Points:**
  - Exchange one currency for another at market rates, speculation on currency values.

### ETFs (Exchange-Traded Funds)

- **Definition:**
  - Pooled investment securities traded like stocks.

- **Key Points:**
  - Basket of securities (stocks, bonds, commodities), intraday trading.

### REITs (Real Estate Investment Trusts)

- **Definition:**
  - Companies owning, operating, or financing income-generating real estate.

- **Key Points:**
  - Publicly traded, steady income from real estate without direct ownership.

### Mutual Funds

- **Definition:**
  - Pooled investment portfolios managed by professionals.

- **Key Points:**
  - Diversified portfolios, purchased at end of trading day, net asset value.

### Hedge Funds

- **Definition:**
  - Actively managed investment pools with diverse trading strategies.

- **Key Points:**
  - Aim for above-average returns, higher fees, often risky strategies.


Long a Call:

Scenario: You're optimistic about a stock's future.
Action: You buy a call option.
Right: To buy the underlying stock at a specified price (strike price).
Analogy: Like making a reservation to buy the stock at a set price.
Short a Call:

Scenario: You own the stock, not expecting a significant price rise.
Action: You sell a call option.
Obligation: Potentially sell your stock at a specified price if the buyer exercises.
Analogy: Offering others the opportunity to buy the stock from you at a certain price.
Long a Put:

Scenario: Concerned about a potential drop in a stock's price.
Action: You buy a put option.
Right: To sell the underlying stock at a specified price (strike price).
Analogy: Having the option to sell the stock at a set price if things go south.
Short a Put:

Scenario: Believing a stock will remain stable or go up.
Action: You sell a put option.
Obligation: Potentially buy the stock at a specified price if the buyer exercises.
Analogy: Giving someone else the opportunity to sell you the stock at a certain price.
