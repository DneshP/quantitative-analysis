**Volume-Weighted Average Price (VWAP)**

The Volume-Weighted Average Price (VWAP) is a trading indicator that calculates the average price of a security weighted by its trading volume. It provides a measure of the average price at which a security has been traded over a specified period, taking into account the volume of trades at each price level.

## Formula

The VWAP is calculated using the following formula:

\[ VWAP = \frac{\sum_{i=1}^{n} \text{Typical Price}_i \times \text{Volume}_i}{\sum_{i=1}^{n} \text{Volume}_i} \]

Where:
- \( n \) is the number of periods (e.g., intraday intervals).
- \(\text{Typical Price}_i = \frac{\text{High}_i + \text{Low}_i + \text{Close}_i}{3}\) is the typical price for period \(i\).
- \(\text{Volume}_i\) is the trading volume for period \(i\).

**Long a Call:**

Scenario: You're optimistic about a stock's future.
Action: You buy a call option.
Right: To buy the underlying stock at a specified price (strike price).
Analogy: Like making a reservation to buy the stock at a set price.

**Short a Call:**

Scenario: You own the stock, not expecting a significant price rise.
Action: You sell a call option.
Obligation: Potentially sell your stock at a specified price if the buyer exercises.
Analogy: Offering others the opportunity to buy the stock from you at a certain price.

**Long a Put:**

Scenario: Concerned about a potential drop in a stock's price.
Action: You buy a put option.
Right: To sell the underlying stock at a specified price (strike price).
Analogy: Having the option to sell the stock at a set price if things go south.

**Short a Put:**

Scenario: Believing a stock will remain stable or go up.
Action: You sell a put option.
Obligation: Potentially buy the stock at a specified price if the buyer exercises.
Analogy: Giving someone else the opportunity to sell you the stock at a certain price.

**Anatomy of a Trade:**

1. **Acquisition of Information and Quotes:**
   - Making informed decisions by gathering relevant information and quotes.

2. **Routing of Order:**
   - Choosing the market and broker as the means to execute the order.

3. **Execution of Order:**
   - Matching the order between the buyer and seller according to the rules of the market.

4. **Confirmation, Clearance, and Settlement:**
   - Confirmation occurs at the end of order execution, ensuring accuracy.
   - Clearance involves recording and matching the trading records.
   - Settlement includes delivering the security and its payment.

**Market Makers:**

# Electronic Market

The electronic market operates on a discrete price grid, with prices linearly arranged based on their magnitude. The price ladder progresses from low to high prices. In any given market, there is always a minimum tick size, representing the adjacent price difference of a financial instrument.

- Stocks generally trade in one-cent tick increments.
- Currencies deal in pips (percentage in point, price interest point).
- Rates deal in bps (basis points).

CDP (Central Depository Participant) acts as the counterparty to both the buyer and seller, assuming the associated risks of both parties.

## Agency Trading

In agency trading, the intermediary facilitates orders between clients without taking positions in the securities being traded. The agency's sole role is to execute client orders, earning commissions or fees for its services.

## Proprietary Trading

In proprietary trading, an entity uses its own capital for market transactions. Unlike agency trading, where the focus is on executing client orders, proprietary trading involves taking speculative positions to capitalize on market movements. Both types of trading are subject to different regulatory conditions.

* Held Order

A held order is where the broker has an obligation to fulfill a client's order.

* Market-Not-Held Order

In a market-not-held order, the institutional trader hires a broker-dealer to execute an order. This provides the broker-dealer with more discretion in the execution process.

**Pegged order**
A Pegged order is employed when a trader aims to maintain a specific distance from the market price of a security. This distance can be applied to both bid and ask prices. For instance, if the current market value of stock XYZ is €50.00 and the trader sets a peg of €0.10, the bid price would be €49.90. The pegged order automatically adjusts its price as the market value of the stock drops or rises, ensuring that the specified distance is maintained.

**Market if touched (MIT)**
Market If Touched (MIT) orders are used to place a bid or ask for a security when a specific price point is touched. The key distinction from buy limit and buy-stop limit orders is that the latter two are executed only if the market price matches the specified level. In contrast, with the MIT order, the price is not guaranteed; it serves as a trigger for a market order. When the trigger price is touched or traded through, the MIT order becomes a market order, executing at the best available market price.
